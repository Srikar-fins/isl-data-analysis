```python
!pip install matplotlib seaborn             
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# FIFA 20 Player Overall Rating Prediction

df = pd.read_csv("C:/Users/srika/OneDrive/Documents/Python/players_20.csv")
print("Shape:", df.shape)
df.head()
```

```python
#Select Relevant Columns
cols = [
    'short_name', 'age', 'height_cm', 'weight_kg', 'nationality',
    'club', 'overall', 'potential', 'value_eur', 'wage_eur',
    'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic',
    'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy',
    'skill_ball_control', 'skill_dribbling', 'movement_agility',
    'movement_balance', 'power_shot_power', 'power_stamina',
    'mentality_vision', 'mentality_composure', 'player_positions'
]
```

```python
#Preprocessing
print("Missing values before:", df.isnull().sum())      
df = df.fillna(df.median(numeric_only=True))

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after:", df.isnull().sum())
df['primary_position'] = df['player_positions'].str.split(',').str[0].str.strip()
df.drop_duplicates(inplace=True)

print("Final shape after preprocessing:", df.shape)
df.head()
```

```python
#Data Visualization
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
df.describe()
```

```python
#Distribution of Overall Rating
plt.figure()
sns.histplot(df['overall'], bins=30, kde=True, color='steelblue')
plt.title('Distribution of Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Number of Players')
plt.show()

#Most Common Positions
top_positions = df['primary_position'].value_counts().head(10)
plt.figure()
sns.barplot(x=top_positions.values, y=top_positions.index, palette='viridis')
plt.title('Top 10 Positions by Player Count')
plt.xlabel('Number of Players')
plt.show()

#Overall Rating by Position
plt.figure()
sns.boxplot(x='primary_position', y='overall', data=df, palette='Set1')
plt.title('Overall Rating by Position')
plt.xticks(rotation=45)
plt.show()

#Age vs Overall Rating
plt.figure()
sns.scatterplot(x='age', y='overall', data=df, alpha=0.4, color='coral')
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(14, 10))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

#Top 10 Features Correlated with Overall Rating ---
corr = numeric_df.corr()['overall'].drop('overall').sort_values(ascending=False)
top_corr = corr.head(10)

plt.figure()
sns.barplot(x=top_corr.values, y=top_corr.index, palette='Blues_r')
plt.title('Top 10 Features Correlated with Overall Rating')
plt.xlabel('Correlation')
plt.show()
```
